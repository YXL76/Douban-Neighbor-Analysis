```python
import requests
from bs4 import BeautifulSoup

from settings import user_agent
from settings import cookie
from settings import target_user

session = requests.Session()

url = 'https://www.douban.com/people/' + target_user + '/rev_contacts'
headers = {
    'User-Agent': user_agent,
    'Cookie': cookie,
}

response = session.get(url=url, headers=headers)
if response.status_code != 200:
    print('获取失败')
soup = BeautifulSoup(response.text, 'lxml')
# print(soup)
print('获取成功')
```

```python
from json import loads

def get_movie_info(uid=''):
    url = 'https://m.douban.com/rexxar/api/v2/user/' + uid + '/collection_stats?type=movie&for_mobile=1&ck=5Kvd'
    referer = 'https://m.douban.com/people/' + uid + '/movie_charts'
    headers = {
        'Referer': referer,
        'User-Agent': user_agent,
    }
    response = session.get(url=url, headers=headers)
    row = []
    decoded = loads(response.text)
    # print(decoded)
    try:
        row.append(decoded['total_collections'])
    except:
        row.append('')
    try:
        row.append(int(decoded['total_spent']))
    except:
        row.append('')
    try:
        row.append(int(decoded['total_cost']))
    except:
        row.append('')
    try:
        row.append(round(decoded['weekly_avg'], 1))
    except:
        row.append('')
    try:
        row.append(decoded['countries'][0]['name'])
    except:
        row.append('')
    try:
        row.append(decoded['countries'][1]['name'])
    except:
        row.append('')
    try:
        row.append(decoded['genres'][0]['name'])
    except:
        row.append('')
    try:
        row.append(decoded['genres'][1]['name'])
    except:
        row.append('')
    # print(row)
    return row


def get_user_info(uid=''):
    url = 'https://m.douban.com/rexxar/api/v2/user/' + uid + '/archives_summary?for_mobile=1&ck=5Kvd'
    referer = 'https://m.douban.com/people/' + uid + '/subject_profile'
    headers = {
        'Referer': referer,
        'User-Agent': user_agent,
    }
    response = session.get(url=url, headers=headers)
    decoded = loads(response.text)
    # print(decoded)
    row = []
    try:
        row.append(decoded['user']['loc']['name'])
    except:
        row.append('')
    try:
        row.append(decoded['user']['statuses_count'])
    except:
        row.append('')
    try:
        row.append(decoded['user']['reg_time'][:4])
    except:
        row.append('')
    try:
        row.append(decoded['user']['gender'])
    except:
        row.append('')
    # print(row)
    return row


def get_info(uid=''):
    row = []
    row += get_user_info(uid)
    row += get_movie_info(uid)
    print(row)
    return row
```

```python
import csv
import numpy as np
import plotly
import plotly.plotly as py
import plotly.graph_objs as go

from settings import username
from settings import api_key
from settings import dataset_file

plotly.tools.set_credentials_file(username=username, api_key=api_key)
```

```python
from settings import loc_lat
from settings import loc_lon
from settings import mapbox_access_token

loc = []
num = []
lat = []
lon = []

with open(dataset_file, 'r', encoding='utf-8') as file:
    csv_file = csv.reader(file)
    for line in csv_file:
        # 如果当前行为空，或者未获取都用户信息（用户已注销）
        if len(line) == 0 or line[0] == '':
            continue
        if loc_lat.get(line[0]) is None:
            continue
        try:
            index = loc.index(line[0])
            # num[index] += 1
        except ValueError:
            loc.append(line[0])
            num.append(1)
            lat.append(loc_lat[line[0]])
            lon.append(loc_lon[line[0]])

# print(loc)
# print(num)            

text = []
index = 0
while index < len(loc):
    text.append(str(loc[index]) + '   ' + str(num[index]))
    index += 1

data = [
    go.Scattermapbox(
        lat=lat,
        lon=lon,
        mode='markers',
        marker=go.scattermapbox.Marker(
            size=9
        ),
        text=text,
    )
]

layout = go.Layout(
    autosize=True,
    hovermode='closest',
    height=800,
    title='',
    mapbox=go.layout.Mapbox(
        accesstoken=mapbox_access_token,
        bearing=0,
        center=go.layout.mapbox.Center(
            lat=34,
            lon=108
        ),
        pitch=0,
        zoom=3.5,
    ),
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='neighbor_distribution_map')
```

```python
from bisect import bisect_left

from settings import status_range

male_status_num = np.array(list(0 for si in status_range))
female_status_num = np.array(list(0 for si in status_range))

with open(dataset_file, 'r', encoding='utf-8') as file:
    csv_file = csv.reader(file)
    for line in csv_file:
        if len(line) == 0 or line[1] == '':
            continue
        if line[3] == 'M':
            index = bisect_left(status_range, int(line[1]))
            male_status_num[index - 1] += 1
        elif line[3] == 'F':
            index = bisect_left(status_range, int(line[1]))
            female_status_num[index - 1] -= 1

# print(male_statuses_num)
# print(female_statuses_num)
            
length = max(max(male_status_num), -max(female_status_num))
# print(length)

boundary = 0
while boundary <= length:
    boundary += 30
# print(boundary)

tickvals = []
ticktext = []

i = 20 - boundary
while i < boundary:
    tickvals.append(i)
    ticktext.append(abs(i))
    i += 20
    
# print(tickvals)
# print(ticktext)

label = []
index = 1
while index < len(status_range):
    label.append(str(status_range[index - 1]) + ' - ' + str(status_range[index]))
    index += 1
label.append(str(status_range[-1]) + '+')
# print(label)

layout = go.Layout(yaxis=go.layout.YAxis(title='Statuse Number'),
                   xaxis=go.layout.XAxis(
                       range=[-boundary, boundary],
                       tickvals=tickvals,
                       ticktext=ticktext,
                       title='Peolpe Number'),
                   barmode='overlay',
                   bargap=0.1)

data = [go.Bar(y=label,
               x=male_status_num,
               orientation='h',
               name='Male',
               hoverinfo='x',
               marker=dict(color='lightskyblue'),
               opacity=0.8
               ),
        go.Bar(y=label,
               x=female_status_num,
               orientation='h',
               name='Female',
               text=-1 * female_status_num.astype('int'),
               hoverinfo='text',
               marker=dict(color='gold'),
               opacity=0.8
               )]

py.iplot(dict(data=data, layout=layout), filename='status_pyramid_chart') 
```

```python
from settings import reg_year_range

reg_year_num = np.array(list(0 for ri in reg_year_range))

with open(dataset_file, 'r', encoding='utf-8') as file: 
    csv_file = csv.reader(file)
    for line in csv_file:
        if len(line) == 0 or line[2] == '':
            continue
        reg_year_num[reg_year_range.index(int(line[2]))] += 1

# print(reg_year_num)

trace = go.Pie(
    labels=reg_year_range,
    values=reg_year_num,
    textinfo='label', 
    marker=dict(line=dict(color='#000000', width=1))
)

py.iplot([trace], filename='reg_year_pie_chart')
```
